Пиксельная маска части изображения

Дано:
- изображение (формат jpg/png)
- диапазон строк для маски (y_min, y_max)
- диапазон столбцов для маски (x_min, x_max)
- число разбиений на области (h, например, для h=10 маска разбивается на 10*10=100 равных областей)

Требуется:
- написать алгоритм усреднения цвета в каждой области (усреднение по каждому цветовому каналу: R,G,B)
- прислать файл(ы): исходное изображение, изображение с маской, код (для работы в формате ipynb - всё в одном файле, в противном случае - все файлы поместить в один архив)
- назвать файл по шаблону на латинице без пробелов: Группа_Фамилия_НомерРаботы (например, 11-000_Ivanov_1)

```python
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
orig_img = imread('https://sun36-1.userapi.com/impg/keb_Sb_GZ0aMyzPW1-CFVDc_M4BdMze9GJbL3g/FhKLz766W1c.jpg?size=736x981&quality=95&sign=82c70cc9065d89365729d0871884e478&type=album')
plt.imshow(orig_img)
plt.show()
```

```python
x_min, x_max = 250, 450
y_min, y_max = 170, 370
h = 10

# Вычислим размеры области
h_width = (x_max - x_min) // h
h_height = (y_max - y_min) // h

edited_img = orig_img.copy()

for x in range(h):
    for y in range(h):
        # вычислим координаты границ области
        x1, y1 = x_min + x * h_width, y_min + y * h_height
        x2, y2 = x1 + h_width, y1 + h_height
        # вычислим усредненный цвет в области
        area = orig_img[y1:y2, x1:x2]
        r_avg = int(np.mean(area[:,:,0]))
        g_avg = int(np.mean(area[:,:,1]))
        b_avg = int(np.mean(area[:,:,2]))
        edited_img[y1:y2, x1:x2, 0] = r_avg
        edited_img[y1:y2, x1:x2, 1] = g_avg
        edited_img[y1:y2, x1:x2, 2] = b_avg

plt.imshow(edited_img)
plt.show()
```
